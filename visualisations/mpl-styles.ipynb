```python
import numpy as np
# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile, Aer, IBMQ
from qiskit.tools.jupyter import *
from qiskit.visualization import *
from ibm_quantum_widgets import *
```

```python
qc = QuantumCircuit(4)
qc.h(1)
qc.rzz(np.pi, 1, 3)
qc.u(np.pi, np.pi, np.pi, 0)
qc.draw('mpl', style={"showindex": True})
```

```python
q_bits = QuantumRegister(4)
cl_bits = ClassicalRegister(3)
pretty_circ = QuantumCircuit(q_bits,cl_bits)

pretty_circ.id(0)
pretty_circ.x(0)
pretty_circ.y(0)
pretty_circ.z(0)
pretty_circ.h(0)
pretty_circ.reset(0)

pretty_circ.s(1)
pretty_circ.sdg(1)
pretty_circ.t(1)
pretty_circ.tdg(1)
pretty_circ.sx(1)
pretty_circ.sxdg(1)

pretty_circ.u(np.pi, np.pi, np.pi, 2)
pretty_circ.r(np.pi, np.pi, 2)
pretty_circ.rx(np.pi, 2)
pretty_circ.ry(np.pi, 2)
pretty_circ.rz(np.pi, 2)
pretty_circ.p(0, 2)

pretty_circ.cx(0,1)
pretty_circ.swap(2,3)
pretty_circ.ccx(0,1,2)
pretty_circ.mcx([0,1, 2], [3])
pretty_circ.cy(0,1)
pretty_circ.cz(2,3)
pretty_circ.dcx(0,1)
pretty_circ.iswap(2,3)
pretty_circ.rxx(np.pi,0,1)
pretty_circ.ryy(np.pi,2,3)
pretty_circ.rzz(np.pi,0,1)

new_circ = QuantumCircuit(2)
new_circ.cx(0,1)
my_gate = new_circ.to_gate(label='my_gate')
pretty_circ.append(my_gate, [2,3])

pretty_circ.barrier()

pretty_circ.measure(0,1)
pretty_circ.draw('mpl')
```
